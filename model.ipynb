```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle
```

```python
sizes = np.random.randint(500, 3500, 100)
```

```python
sizes
```

```python
bedrooms=np.random.randint(1,6,100)
```

```python
bedrooms
```

```python
prices=(sizes*0.3) + (bedrooms * 50) +np.random.randint(-200,200,100)
```

```python
prices
```

```python
data=pd.DataFrame(
    {
        "size_sqft":sizes,
        "bedrooms":bedrooms,
        "price":prices       
    }
)
```

```python
data.head()
```

```python
x=data[["size_sqft","bedrooms"]]
```

```python
y=data["price"]
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
sample_X= np.array([[200,3],[1500,2]])
```

```python
model.predict(sample_X)
```

```python
with open("house_price_model.pkl","wb") as f:
     pickle.dump(model,f)
```
